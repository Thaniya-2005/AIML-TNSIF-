TASK - Iris Dataset Visualization

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="petal_length", y="petal_width", data=iris)
plt.title("Scatter Plot for Petal Length and Width")
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris, hue="species")
plt.title("Color points based on species")
```

```python
plt.figure(figsize=(6,5))
sns.pairplot(iris, hue="species")
plt.suptitle("Pairplot",y=1.02)
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species")
plt.title('Species Separation via Petal Dimensions')
```

Practice With Different Build In - Penguines

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=penguins)
plt.title("Scatter Plot")
```

```python
plt.figure(figsize=(6,6))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins, hue="sex")
plt.title("Line Plot")
```

```python
plt.figure(figsize=(5,5))
sns.barplot(y="body_mass_g", data=penguins, hue="sex")
plt.title("BarPlot")
```

```python
plt.figure(figsize=(3,5))
sns.countplot(y="island", data=penguins, hue="sex")
plt.title("CountPlot - ISLAND")
plt.show()

plt.figure(figsize=(5,3))
sns.countplot(x="species", data=penguins, hue="sex")
plt.title("CountPlot - SPECIES")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x="body_mass_g", y="island", data=penguins, hue="sex")
plt.title("Box Plot - BodyMass")
```

```python
plt.figure(figsize=(6,7))
sns.violinplot(x="flipper_length_mm", data=penguins, hue="sex")
plt.title("Violin Plot")
```

```python
plt.figure(figsize=(5,5))
corr = penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="PiYG", linewidth=0.7)
#sns.heatmap(corr, annot=True, cmap="BrBG", linewidth=0.7)
#sns.heatmap(corr, annot=True, cmap="RdBu", linewidth=0.7)
#sns.heatmap(corr, annot=True, cmap="seismic", linewidth=0.7)
#sns.heatmap(corr, annot=True, cmap="bwr", linewidth=0.7)
#sns.heatmap(corr, annot=True, cmap="coolwarm", linewidth=0.7)
plt.title("HeatMap")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.pairplot(penguins, hue="sex", )
plt.title("Pair Plot")
```

```python
plt.figure(figsize=(4,5))
sns.histplot(penguins["body_mass_g"], kde=True, bins=20)
plt.title("Distibution of Body Mass")
```

```python
plt.figure(figsize=(9,9))
sns.regplot(x="body_mass_g", y="flipper_length_mm", data=penguins)
plt.title("Regression Plot")
```
