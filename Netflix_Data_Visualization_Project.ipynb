```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
data = df.dropna()
data.head()
```

```python
type_counts = data['type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Content Types on Netflix')
plt.show()
```

```python
countries_df = data['country'].dropna().str.split(', ').explode()
top_countries = countries_df.value_counts().head(15)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='turbo')
plt.title('Top 15 Contunries Contributing Content to Netflix')
plt.show()
```

```python
directors_df = data['director'].dropna().str.split(', ').explode()
top_directors = directors_df.value_counts().head(15)
plt.figure(figsize=(15, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, hue=top_directors.index, palette='tab20')
plt.title('Top 15 Directors with the Most Titles on Netflix')
plt.show()
```
